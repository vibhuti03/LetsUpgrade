```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

## Selecting Independent and dependent features

```python
X = df.drop('diagnosis',axis=1)
X.head()
```

```python
y = df['diagnosis']
y.head()
```

## Separating training data and testing data
<br>70% and 30%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
X_train.head()
```

## Scaling the independent variables

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
#fit -> study
#transform -> scale it into a range centered around 0
```

# Machine Learning Models

```python
results = pd.DataFrame()
```

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
lr_acc = accuracy_score(y_test, y_pred)
print(lr_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression'], 'Accuracy':[lr_acc]})
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy']]
results
```

## 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc =DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
dtc_acc = accuracy_score(y_test, y_pred)
print(dtc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'], 'Accuracy':[dtc_acc]})
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy']]
results
```

## 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
rfc_acc = accuracy_score(y_test, y_pred)
print(rfc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Random Forest Classifier'], 'Accuracy':[rfc_acc]})
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy']]
results
```

## 4. Support Vector Classifier

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
svc_acc = accuracy_score(y_test, y_pred)
print(svc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Support Vector Classifier'], 'Accuracy':[svc_acc]})
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy']]
results
```
